```python
import plotly
import plotly.plotly as py
import pandas as pd
import numpy as np
import matplotlib 
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')
import plotly.graph_objs as go

%matplotlib inline

pd.set_option('display.max_rows', 10)
pd.set_option('display.max_columns', 15)

# path to PROJECT_DATA.xslx saved on local directory
path = r....\\PROJECT_DATA.xlsx"
```

# Graph 1- Distribution of national income among population for several countries

```python
# graph1
df1 = pd.read_excel(io = path, sheet_name = 'GRAPH1')
df1

names = df1['Country']
trace1 = go.Bar(
    x=df1['Country'],
    y=df1['p0p50'],
    text= df1['p0p50'],
    textposition = 'auto',
    name='Bottom (0-50)',
    width = 0.6,
    marker=dict(
        color='rgb(204,204,204)', 
    )
)
trace2 = go.Bar(
    x=df1['Country'],
    y=df1['p50p90'],
    text=df1['p50p90'],
    textposition = 'auto',
    name='Middle (50-90)',
    width = 0.6,
    marker=dict(
        color='rgb(51,153,255)'
    )
)

trace3 = go.Bar(
    x=df1['Country'],
    y=df1['Top10_split'],
    text= df1['Top10_split'],
    textposition = 'auto',
    name='Top 9% (90-99)',
    width = 0.6,
    marker=dict(
        color='rgb(153,51,255)'
    )
)

trace4 = go.Bar(
    x=df1['Country'],
    y=df1['p99p100'],
    text= df1['p99p100'],
    textposition = 'auto',
    name='Top 1% (99-100)',
    width = 0.6,
    marker=dict(
        color='rgb(255,153,51)'
        ),
    opacity=1
)

data = [trace1, trace2, trace3, trace4]
layout = go.Layout(
    barmode='stack', 
    title='Distribution of National Income Among Population',
    titlefont = dict(
        size = 30,
    ),
    xaxis=dict(
        titlefont= dict(
        size = 18),
        tickfont=dict(
            size=18,
            color='rgb(107, 107, 107)'
        )
    ),
    yaxis=dict(
        title='% Income',
        titlefont=dict(
            size=18,
            color='rgb(107, 107, 107)'
        ),
        tickfont=dict(
            size=14,
            color='rgb(107, 107, 107)'
        )
    ),

    legend=dict(
        x=25,
        y=1,
        orientation = "v",
        bgcolor='rgba(255, 255, 255, 0)',
        bordercolor='rgba(255, 255, 255, 0)'
    )
)

fig = go.Figure(data=data, layout=layout)
plotly.offline.plot(fig, filename='Graph1')
```

# Graph 9-Assets by quintile 2016

```python
# graph9
df9 = pd.read_excel(io = path, sheet_name = 'GRAPH9', index_col = 0)

df9

first = df9['Lowest quintile'].as_matrix()
second = df9['Second quintile'].as_matrix()
third = df9['Middle quintile'].as_matrix()
fourth = df9['Fourth quintile'].as_matrix()
fifth = df9['Highest quintile'].as_matrix()

data = np.c_[first, second, third, fourth, fifth]

import plotly.plotly as py
import plotly.graph_objs as go

trace = go.Heatmap(z = data,
                   x = ['Lowest', 'Second', 'Middle', 'Fourth', 'Highest'],
                  y = list(df9.index), 
                  colorscale = 'Viridis')
data=[trace]
layout = dict( margin = dict(t=100,r=20,b=30,l=400),
     width = 1000, height = 700,         
    autosize = False, 
    title = 'Assets by Qunitile 2016',
    titlefont = dict(
              size = 24,))


figure = dict(data=data,layout=layout)
plotly.offline.plot(figure, filename='Graph9')
```
